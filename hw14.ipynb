```python
#TASK 1.1
n=20
for i in range(n):
    i*=i  
    print(i)
    if i>=n:
        break

#TC = O(n)  - no nested loops, it runs proportionately to the n
```

```python
#1.2
for i in range(n):
    i=pow(2,i) 
    print(i)
    if i>(n+1):
        break
#TC = O(log(n))  - i grows faster then n, so runs log(n) times
```

```python
#2.1
def digit_sum(n):
    if n > 10:
        return (n % 10) + digit_sum(n // 10) #sums th remaider part (from the tail)
    else:                                    # with the beginning numbers
        return n
        
```

```python
digit_sum(298558)
```

```python
#2.2
def ap_sum(num):
    if num > 1:
        return  num + ap_sum(num-1)
    else:
        return num
```

```python

ap_sum(9)
```

```python
#2.3 To be fair I came across this algortihm when I was reading about 
#them during the last homework, so it wasn't difficult to write it
def fibonacci(n):
    if n <= 1:
        return n
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)
```

```python
print(fibonacci(6),
fibonacci(9))
```

But understanding the TC for fibonnacci was quite difficult: each iteration creates 2 branches (for fibonacci(n-1) and fibonacci(n-2)), or takes it to the power of n

So the time complexity is O(2^n) - exponential
